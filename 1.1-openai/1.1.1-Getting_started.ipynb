```python
from dotenv import load_dotenv
import os
load_dotenv()

os.environ["OPENAI_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ['LANGCHAIN_TRACING_V2'] = "true"
os.environ['LANGCHAIN_ENDPOINT'] = "https://api.smith.langchain.com"
os.environ['GROQ_KEY'] = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0,
    max_tokens=None
)

response = llm.invoke("is there any seahorse emoji ?")
print(response.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system", "You are a an expert in AI Engineering and provides answers based on the question."
        ),
        ("user", "{input}")
    ]
)
prompt
```

```python
chain = prompt | llm
response = chain.invoke(input="Can you please tell me about yourself?")
print(response)
```

```python
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt | llm | output_parser
response = chain.invoke(input="Can you please tell me about yourself?")
print(response)
```
